## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
df_mat
````

````python
total_questoes = {'Matemática':24,
                  'Português':18,
                  'Geografia':8,
                  'Inglês':8,
                  'História':8,
                  'Física':12,
                  'Química':12
                 }
````

````python
# 1) Seu código aqui
#Matemática
acerto_mat = df_mat['Qt_acertos'] * 100 / total_questoes['Matemática']
df_mat['Porcentual'] = acerto_mat

#Português
acerto_por = df_por['Qt_acertos'] * 100 / total_questoes['Português']
df_mat['Porcentual'] = acerto_por

#Geografia
acerto_geo = df_geo['Qt_acertos'] * 100 / total_questoes['Geografia']
df_geo['Porcentual'] = acerto_geo

#Inglês
acerto_ing = df_ing['Qt_acertos'] * 100 / total_questoes['Inglês']
df_ing['Porcentual'] = acerto_ing

#História
acerto_his = df_his['Qt_acertos'] * 100 / total_questoes['História']
df_his['Porcentual'] = acerto_his

#Física
acerto_fis = df_fis['Qt_acertos'] * 100 / total_questoes['Física']
df_fis['Porcentual'] = acerto_fis

#Química
acerto_qui = df_qui['Qt_acertos'] * 100 / total_questoes['Química']
df_qui['Porcentual'] = acerto_qui

df_mat
````

````python
# 2) Seu código aqui
#Matemática
df_mat['Total Acertos'] = df_mat['Qt_acertos'] + df_por['Qt_acertos'] + df_geo['Qt_acertos'] + df_ing['Qt_acertos'] + df_his['Qt_acertos'] + df_fis['Qt_acertos'] + df_qui['Qt_acertos'] 

#Português
df_por['Total Acertos'] = df_mat['Qt_acertos'] + df_por['Qt_acertos'] + df_geo['Qt_acertos'] + df_ing['Qt_acertos'] + df_his['Qt_acertos'] + df_fis['Qt_acertos'] + df_qui['Qt_acertos']

#Geografia
df_geo['Total Acertos'] = df_mat['Qt_acertos'] + df_por['Qt_acertos'] + df_geo['Qt_acertos'] + df_ing['Qt_acertos'] + df_his['Qt_acertos'] + df_fis['Qt_acertos'] + df_qui['Qt_acertos']

#Inglês
df_ing['Total Acertos'] = df_mat['Qt_acertos'] + df_por['Qt_acertos'] + df_geo['Qt_acertos'] + df_ing['Qt_acertos'] + df_his['Qt_acertos'] + df_fis['Qt_acertos'] + df_qui['Qt_acertos']

#História
df_his['Total Acertos'] = df_mat['Qt_acertos'] + df_por['Qt_acertos'] + df_geo['Qt_acertos'] + df_ing['Qt_acertos'] + df_his['Qt_acertos'] + df_fis['Qt_acertos'] + df_qui['Qt_acertos']

#Física
df_fis['Total Acertos'] = df_mat['Qt_acertos'] + df_por['Qt_acertos'] + df_geo['Qt_acertos'] + df_ing['Qt_acertos'] + df_his['Qt_acertos'] + df_fis['Qt_acertos'] + df_qui['Qt_acertos']

#Química
df_qui['Total Acertos'] = df_mat['Qt_acertos'] + df_por['Qt_acertos'] + df_geo['Qt_acertos'] + df_ing['Qt_acertos'] + df_his['Qt_acertos'] + df_fis['Qt_acertos'] + df_qui['Qt_acertos']
````

````python
# 3) Seu código aqui
n_questoes = total_questoes['Matemática'] + total_questoes['Português'] + total_questoes['Geografia'] + total_questoes['Inglês'] + total_questoes['História'] + total_questoes['Física'] + total_questoes['Química']

#Matemática
porcentagem_geral = df_mat['Total Acertos'] * 100 / n_questoes
df_mat['Porcentual Geral'] = porcentagem_geral

#Português
porcentagem_geral = df_por['Total Acertos'] * 100 / n_questoes
df_por['Porcentual Geral'] = porcentagem_geral

#Geografia
porcentagem_geral = df_geo['Total Acertos'] * 100 / n_questoes
df_geo['Porcentual Geral'] = porcentagem_geral

#Inglês
porcentagem_geral = df_ing['Total Acertos'] * 100 / n_questoes
df_ing['Porcentual Geral'] = porcentagem_geral

#História
porcentagem_geral = df_his['Total Acertos'] * 100 / n_questoes
df_his['Porcentual Geral'] = porcentagem_geral

#Física
porcentagem_geral = df_fis['Total Acertos'] * 100 / n_questoes
df_fis['Porcentual Geral'] = porcentagem_geral

#Química
porcentagem_geral = df_qui['Total Acertos'] * 100 / n_questoes
df_qui['Porcentual Geral'] = porcentagem_geral
````

````python
# 4) Seu código aqui
nota_corte = 45
total_aprovados = 0 

for aluno, linha in df_mat.iterrows():
    if linha['Total Acertos'] >= nota_corte: 
        total_aprovados += 1
    else:
        df_mat = df_mat.drop(aluno)
        df_por = df_por.drop(aluno)
        df_geo = df_geo.drop(aluno)
        df_ing = df_ing.drop(aluno)
        df_his = df_his.drop(aluno)
        df_fis = df_fis.drop(aluno)
        df_qui = df_qui.drop(aluno)
        
print(total_aprovados)
df_mat
````
